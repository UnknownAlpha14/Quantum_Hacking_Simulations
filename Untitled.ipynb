```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.providers.aer import QasmSimulator
import random

# Parameters
num_qubits = 100
num_intercepts = 10

# Execute the intercept-resend attack multiple times
simulator = QasmSimulator()

for intercept in range(num_intercepts):

    # Create a quantum circuit
    qc = QuantumCircuit(num_qubits, num_qubits)

    alice_bits = [random.choice([0,1]) for _ in range(num_qubits)]
    alice_bases = [random.choice(['Z', 'X']) for _ in range(num_qubits)]
    eve_bases = [random.choice(['Z', 'X']) for _ in range(num_qubits)]
    bob_bases = [random.choice(['Z', 'X']) for _ in range(num_qubits)]

    for i, bit in enumerate(alice_bits):
        if alice_bases[i] == 'X':
            qc.h(i)
        if bit:
            qc.x(i)

    # Eve intercepts and tries to measure
    for i, base in enumerate(eve_bases):
        if base == 'X':
            qc.h(i)
    qc.measure_all()

    # Reset the qubits to prepare for Bob's measurement
    qc.reset(qc.qubits)

    # Alice re-encodes her qubits based on her bits and bases
    for i, bit in enumerate(alice_bits):
        if alice_bases[i] == 'X':
            qc.h(i)
        if bit:
            qc.x(i)

    # Bob measures the qubits
    for i, base in enumerate(bob_bases):
        if base == 'X':
            qc.h(i)
    qc.measure_all()

    # Execute the circuit
    compiled_circuit = transpile(qc, simulator)
    result = simulator.run(compiled_circuit).result()
    counts = result.get_counts(qc)

    # Extract Bob's received bits from the counts 
    bob_bits_string = list(counts.keys())[0]
    bob_bits = [int(bit) for bit in bob_bits_string.split(' ') if bit]

    # Compare Alice's sent bits with Bob's received bits in the same basis
    errors = sum([1 for a_bit, a_base, b_bit, b_base in zip(alice_bits, alice_bases, bob_bits, bob_bases) 
                  if a_base == b_base and a_bit != b_bit])
    same_basis = sum([1 for a_base, b_base in zip(alice_bases, bob_bases) if a_base == b_base])

    error_rate = errors / same_basis
    print(f"Iteration {intercept + 1}: Error Rate: {error_rate*100:.2f}%")

print("All iterations completed.")
```
